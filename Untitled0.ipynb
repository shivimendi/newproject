```python
!pip install shap xgboost lightgbm scikit-learn
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
import xgboost as xgb
import lightgbm as lgb
from sklearn.svm import SVR
import shap

# Load your data
data = pd.read_excel('DataForClassificationAllFilled.xlsx')
X = data.iloc[:, :-1]  # Predictor variables
y = data.iloc[:, -1]   # Response variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define models
models = {
    'RandomForest': RandomForestRegressor(),
    'XGBoost': xgb.XGBRegressor(),
    'LightGBM': lgb.LGBMRegressor(),
    'GradientBoosting': GradientBoostingRegressor(),
    'SVM': SVR()
}

best_model = None
best_model_name = ''
best_score = float('inf')

# Evaluate models using cross-validation
for name, model in models.items():
    scores = cross_val_score(model, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
    mean_score = -scores.mean()

    print(f'{name} MSE: {mean_score:.4f}')

    if mean_score < best_score:
        best_score = mean_score
        best_model = model
        best_model_name = name

# Fit the best model on the training data
best_model.fit(X_train, y_train)

if best_model_name in ['RandomForest', 'XGBoost', 'LightGBM', 'GradientBoosting']:
    explainer = shap.TreeExplainer(best_model)
elif best_model_name == 'SVM':
    explainer = shap.KernelExplainer(best_model.predict, X_train)
else:
    raise ValueError("Unsupported model type for SHAP.")

# Calculate SHAP values
shap_values = explainer(X_train)

# Calculate mean absolute SHAP values for feature importance
shap_importance = np.abs(shap_values.values).mean(axis=0)

# Create a DataFrame to display feature importance
feature_names = X.columns
shap_importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': shap_importance
})

# Sort the DataFrame by importance
shap_importance_df = shap_importance_df.sort_values(by='Importance', ascending=False)

# Display the best model and feature importance
print(f'Best Model: {best_model_name}')
print(shap_importance_df)

# Optional: Plot SHAP values for feature importance
shap.summary_plot(shap_values, X_train, feature_names=feature_names)
```
